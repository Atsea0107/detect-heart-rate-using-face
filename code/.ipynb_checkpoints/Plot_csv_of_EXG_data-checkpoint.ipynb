```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = 'data'
file_name = 'Part_1_S_Trial1_emotion.csv'

data = pd.read_csv(os.path.join(base, file_name), index_col=0)
```

```python
start = 0
end = 5000

fig, ax = plt.subplots(3, 1, figsize=(16,16), 
                       gridspec_kw = {'wspace':0.2, 'hspace':0.8})

for index, number in enumerate(range(1, 4)):
    name = "EXG" + str(number)
    temp_data = data[name][start:end]
    
    ax[index].title.set_text(name)
    ax[index].set_ylim((min(temp_data), max(temp_data)))
    ax[index].plot(temp_data)
    
plt.show()
```
